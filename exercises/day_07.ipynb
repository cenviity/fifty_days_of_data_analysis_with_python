# Day 7: Transpose and Swap Arrays

```python
import numpy as np
```

## Q1

```python
# Create array of numbers
arr1 = np.arange(0, 10).reshape((2, 5))
arr1
```

## Q2

```python
# Change the shape of the array
arr2 = arr1.reshape((5, 2))
arr2
```

```python
# Transpose the second array and add the two arrays together
arr3 = arr1 + arr2.T
arr3
```

## Q3

```python
# Swap the indexes of the array above
arr4 = np.swapaxes(arr3, 0, 1)
arr4
```

## Q4

```python
# Slice row 1 from the array above
arr4[0]
```

## Q5

```python
players = ["Robin", "Leo", "Pogba", "Diego", "Ronaldo"]
teams = ["Man UTD", "Barca", "Juve", "Napo", "RMadrid"]

goals = [
    [12, 15, 16, 15, 13],
    [26, 30, 31, 25, 24],
    [10, 12, 8, 6, 13],
    [18, 19, 17, 20, 21],
    [21, 32, 25, 21, 22],
]
```

```python
# Define season numbers
seasons = np.arange(1, 6)
seasons
```

```python
# Create heatmap of the goals
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

# Set colour gradient scheme and add grid labels
sns.heatmap(goals, cmap="Oranges", annot=True)

# Add axis labels
ax.set_xlabel("Season")
ax.set_ylabel("Player")

# Add tick labels for seasons
ax.set_xticks(range(len(seasons)), seasons)
# Add tick labels for players and their respective teams
player_team_names = ["\n".join((player, team))
                     for player, team in zip(players, teams)]
ax.set_yticks(range(len(players)), player_team_names)

# Remove tick marks
ax.tick_params(bottom=None, left=None)

# Add title
plt.title("Goals scored by players over five seasons")
plt.show()
```
