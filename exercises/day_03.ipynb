# Day 3: Generating Random Arrays

```python
import numpy as np
```

## Q1

```python
# Create an array of random numbers
rng = np.random.default_rng(seed=7)
arr = rng.random((3, 4))
arr
```

## Q2

```python
# Create an array of random integers
rng = np.random.default_rng(seed=7)
arr = rng.integers(0, 10, (3, 4))
arr
```

## Q3

```python
# Create an array of 1000 random numbers from a standard normal distribution
rng = np.random.default_rng(seed=7)
arr = rng.standard_normal(1000)
arr
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.figure(figsize=(10, 8))
plt.hist(arr, bins=20, color="orange")
plt.title("Standard normal distribution", fontsize=15)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
